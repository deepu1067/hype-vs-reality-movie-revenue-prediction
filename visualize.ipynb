```python
import seaborn as sns
import matplotlib.pyplot as plt
df = df.groupby('genre')[['production budget', 'worldwide gross']].mean().reset_index()


sns.scatterplot(data=df, x='production budget', y='worldwide gross', alpha=0.7)
plt.title('Production Budget vs Worldwide Gross by Genre')
plt.show()
```

```python
import pandas as pd
genres_split = df['genre'].str.split(', ')
all_genres = [genre for sublist in genres_split for genre in sublist]

# Step 3: Count Each Genre
genre_counts = pd.Series(all_genres).value_counts()

# Step 4: Plot the Genre Distribution
plt.figure(figsize=(12, 8))
sns.barplot(x=genre_counts.values, y=genre_counts.index, palette='viridis')
plt.title('Number of Movies by Genre')
plt.xlabel('Count')
plt.ylabel('Genre')
plt.tight_layout()
plt.show()

plt.savefig(f'genre.png', dpi=300, bbox_inches='tight')
```
